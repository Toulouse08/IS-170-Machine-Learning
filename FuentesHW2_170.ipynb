```python
import pandas as pd
import numpy as np

df=pd.read_csv("Summer_Sports_Experience.csv")
df
```

```python
df['Sunday\'s Attendance']=df['Sunday\'s Attendance'].replace(0,df['Sunday\'s Attendance'].mean())
df
```

```python
df["Monday\'s Attendance"]=df["Monday's Attendance"].replace(0,df["Monday's Attendance"].mean())
df
```

```python
df["Thursday's Attendance"]=df["Thursday's Attendance"].replace(0,df["Thursday's Attendance"].mean())
df
```

```python
df["Friday's Attendance"]=df["Friday's Attendance"].replace(0,df["Friday's Attendance"].mean())
df
```

```python
df["Saturday's Attendance"]=df["Saturday's Attendance"].replace(0,df["Saturday's Attendance"].mean())
df
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
label = le.fit_transform(df['Attendance Sum'])
label
```

```python
df.drop("Attendance Sum", axis=1, inplace=True)
df["Attendance Sum"] = label
df
```

```python
old_cols=df.columns.values
new_cols=["Sports Played", "Borough Location", "Park Location", "Week Start Date", "Sunday's Attendance", "Monday's Attendance", "Tuseday's Attendance", "Wednesday's Attendance", "Thursday's Attendance", "Friday's Attendance", "Saturday's Attendance", "Attendance Sum"]
df=df.reindex(columns=new_cols)
df
```
