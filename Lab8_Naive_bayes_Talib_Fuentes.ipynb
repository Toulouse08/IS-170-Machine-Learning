#Naive-Bayes with wine Data

```python
from sklearn import datasets
import pandas as pd

#load dataset
wine=datasets.load_wine()
```

```python
df=pd.read_csv("wine.csv")
df.head()
```

```python
df.columns= ['class', 'alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash', 'magnesium', 'total_phenols', 'flavanoids', 'nonflavanoid_phenols', 'proanthocyanins', 'color_intensity', 'hue', 'od280/od315_of_diluted_wines', 'proline',]
df.head()
```

```python
x_data = df[['alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash', 'magnesium', 'total_phenols', 'flavanoids', 'nonflavanoid_phenols', 'proanthocyanins', 'color_intensity', 'hue', 'od280/od315_of_diluted_wines', 'proline']]
y_data = df['class']
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3) # 70% training and 30% test
```

```python
#This is model generation

#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

#Train the model using the training sets
model.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)

#This is to evaluate the model
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#In order to process we must turn those categorical values into numberic values
#Import LabelEncoder
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()

#Using .fit_transform function to fit label
#encoder and return encoded label
gender_label=le.fit_transform(df1['gender'])

#printing gender_label
gender_label
```

```python


#Using .fit_transform function to fit label
#encoder and return encoded label
married_label=le.fit_transform(df1['ever_married'])

#printing gender_label
married_label
```

```python


#Using .fit_transform function to fit label
#encoder and return encoded label
smoking_label=le.fit_transform(df1['smoking_status'])

#printing gender_label
smoking_label
```

```python
#we'll use zip method in this function called merge to combine the two labels: weather and temp
def merge (list1, list2, list3, list4, list5, list6, list7, list8):
  merged_list=tuple(zip(list1, list2, list3, list4, list5, list6, list7, list8))
  return merged_list

features=merge(smoking_label, married_label, age_label, gender_label, hypertension_label, heart_label, glucose_label, bmi_label)
features
```

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

# Train the model using the training sets
model.fit(features,stroke_target)

#Predict Output
predicted= model.predict([[0,1]]) # 0:Overcast, 2:Mild
print ("Predicted Value:", predicted)
```

```python
x_data = features
y_data = stroke_target
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3,random_state=109) # 70% training and 30% test
```

```python
#This is model generation

#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

#Train the model using the training sets
model.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)

#This is to evaluate the model
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
